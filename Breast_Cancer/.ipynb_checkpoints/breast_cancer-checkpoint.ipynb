### I. Hiểu về Vấn đề phân tích Ung thư vú
- Ung thư vú (Breast cancer) là một trong những bệnh ung thư phổ biến nhất xảy ra ở phụ nữ trên thế giới.
- Nguy cơ ung thư vú ở phụ nữ mỗi nước tuy có khác nhau nhưng tỷ lệ mắc bệnh trung bình khá cao . Ví dụ, nguy cơ trung bình của 1 phụ nữ ở Mỹ phát triển ung thư vú vào 1 thời điểm nào đó trong đời là khoảng 13%, nói cách khác, cứ 8 người thì có 1 người có khả năng bị ung thư vú! Hàng năm, trên thế giới có khoảng 2 triệu người được chẩn đoán mắc ung thư vú và khoảng người 600.000 người tử vong vì căn bệnh này.



### II. Phân tích dữ liệu thăm dò (EDA - Exploratory Data Analysis)

#### 1. Lấy dữ liệu
- Tải file data.csv từ Kaggle. File chứa dữ liệu về các thông tin liên quan đến Khối u (để chẩn đoán ung thư) về hình dạng, kích thước. 
- Khối u được phân loại là lành tính hoặc ác tính dựa trên hình dạng và kích thước của nó. Các đặc điểm được tính toán từ hình ảnh số hóa được lấy từ quá trình Sinh thiết vú (breast-biopsy link đọc tham khảo thêm https://www.mountelizabeth.com.sg/vi/healthplus/article/guide-breast-biopsy). Nó mô tả các đặc điểm của nhân tế bào có trong hình ảnh.
- Các features (đặc tính) của tập dữ liệu mà chúng ta cần biết:
    1. Bán kính khối u (tumor radius - k/cách từ tâm đến điểm ở chu vi khối u)
    2. Kết cấu (texture - 
    3. Chu vi (perimeter)
    4. Diện tích (area)
    5. Độ mịn (smoothess - sự thay đổi cục bộ trong phạm vi độ dài)
    6. Mật độ (compactness - perimeter² / area — 1.0)
    7. Độ cong (concavity - mức độ nghiêm trọng của các phần lõm của đường viền)
    8. Điểm lõm (concave points - Số phần lõm của đường viền)
    9. Cân đối (symmetry)
    10. Kích thước Fractal (Fractal dimension - mô tả sự phát triển mẫu khối u là 1 số liên tục từ 1 -> 2. Số này thể hiện mức độ phức tạp của kiến trúc mô của mẫu sinh thiết, càng cao thì mô càng bất thường.)
    

#### 2. Phân tích dữ liệu
- Thực hiện phân tích bằng ngôn ngữ python, sử dụng các thư viện hỗ trợ như pandas, matplotlib, ...


```python
import numpy as np # linear algebra
import pandas as pd # data processing
import seaborn as sns # data visualization
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('data.csv')

# Đầu tiên chúng ta bắt đầu xem các features của dữ liệu
data.head() # head() method show only first 5 rows of data
```

```python
data.describe()
```

```python
# features names as a list
col = data.columns
print(col)
```

Ta thấy, dataset có 569 dòng (rows), 33 cột (columns). Có 2 cột ta có thể lược bỏ đi là cột **id** và cột **Unnamed:32** đều chứa NaN không cần thiết.Cột **Diagnosis** (Chẩn đoán) chứa nhãn Lành tính (B- Benign) hay Ác tính (M- Malignant)

```python
# y includes labels, x include features
y = data.diagnosis

drop_col = ['id', 'Unnamed: 32', 'diagnosis']
x = data.drop(drop_col, axis = 1)
x.head()
```

```python
data['diagnosis'].value_counts(normalize=True)
ax = sns.countplot(y, label="Count")
B, M = y.value_counts()
print('Số case Lành tính: ', B)
print('Số case Ác tính: ', M)
```

Từ biểu đồ ta có thể thấy được số lượng mẫu bệnh nhân ác tính khá thấp so với số lượng mẫu bênh nhân lành tính.

```python
data_dia = y
data = x
data_n_2 = (data - data.mean()) / (data.std())              # standardization
data = pd.concat([y,data_n_2.iloc[:,0:10]],axis=1)
data = pd.melt(data,id_vars="diagnosis",
                    var_name="features",
                    value_name='value')
plt.figure(figsize=(10,10))
sns.violinplot(x="features", y="value", hue="diagnosis", data=data,split=True, inner="quart")
plt.xticks(rotation=90)
```
